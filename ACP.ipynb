```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Chargement
data = load_iris()

X = data.data

# 2. Centrage et Réduction (Crucial !)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Application de l'ACP
pca = PCA(n_components=2)
components = pca.fit_transform(X_scaled)

# 4. Visualisation
plt.figure(figsize=(8,6))
plt.scatter(components[:, 0], components[:, 1], c=data.target)
plt.xlabel('Composante Principale 1')
plt.ylabel('Composante Principale 2')
plt.title('Projection ACP du dataset Iris')
plt.show()
```

### Réduction de dimension : On passe de 4 variables à 2 variables sans perdre l'essentiel de l'info.

### Visualisation : On peut enfin "voir" les données en 2D alors qu'elles sont en 4D.

### Préparation au Clustering : C'est la base avant d'attaquer des algorithmes comme les K-Means.
